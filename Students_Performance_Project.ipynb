```python
numerical = ["math score", "reading score", "writing score"]

def plot(var):
    print(sb.displot(x = data[var], data = data, hue = "gender", multiple = "stack"))

for i in numerical: 
    plot(i)
```

```python
# 0 : fail  1 : pass
passMark = 50
data["Math_passStatus"] = np.where(data["math score"] < passMark, 0, 1)
data["Writing_passStatus"] = np.where(data["writing score"] < passMark, 0, 1)
data["Reading_passStatus"] = np.where(data["reading score"] < passMark, 0, 1)
```

```python
data.head()
```

```python
passStatus = ["Math_passStatus", "Writing_passStatus", "Reading_passStatus"]

for i in passStatus:
    print(data[i].value_counts())
    print("----------")
```

```python
passStatus = ["Math_passStatus", "Writing_passStatus", "Reading_passStatus"]

def plot(var):
    print(sb.displot(y = "parental level of education", data = data, hue = data[var], col = data[var]))

for i in passStatus: 
    plot(i)
```

```python
data["OverAll_PassStatus"] = data.apply(lambda x: 0 if x["Math_passStatus"] == 0 or x["Reading_passStatus"] == 0 or x["Writing_passStatus"] == 0 else 1, axis = 1)
data.OverAll_PassStatus.value_counts()
```

```python
sb.catplot(y = "parental level of education", data = data, kind = "count", hue = "OverAll_PassStatus");
```

```python
data.head()
```

```python
data["Average_Of_Scores"] = (data["math score"] + data["reading score"] + data["writing score"]) / 3
```

```python
data.head()
```

```python
sb.catplot(y = "Average_Of_Scores", data = data, kind = "count", height = 20);
```

```python
data.nunique()
```

```python
def Grade(OverAll_PassStatus, Average_Of_Scores):
    
    if(OverAll_PassStatus == 0):
        return "F"
    if(Average_Of_Scores >= 90):
        return "A"
    if(Average_Of_Scores >= 80):
        return "B"
    if(Average_Of_Scores >= 70):
        return "C"
    if(Average_Of_Scores >= 60):
        return "D"
    if(Average_Of_Scores >= 50):
        return "E"
    else:
        return "F"
    
data["Grade"] = data.apply(lambda x: Grade(x["OverAll_PassStatus"], x["Average_Of_Scores"]), axis = 1)  

data.Grade.value_counts()
```

```python
sb.catplot(x = "Grade", data = data, kind = "count");
```

```python
sb.catplot(x = "Grade", data = data, hue = "parental level of education", kind = "count");
```

```python
male_count = data["gender"].value_counts()["male"]
female_count = data["gender"].value_counts()["female"]

male_ratio = male_count / (female_count + male_count) * 100
female_ratio = female_count / (female_count + male_count) * 100

print(f"Female Ratio : {female_ratio}\nMale Ratio : {male_ratio}")
```

```python
labels = ["Female","Male"]
colors = ["pink", "lightblue"]
sizes = [female_ratio, male_ratio]
plt.pie(sizes, colors = colors, labels = labels, startangle = 90, autopct = "%1.1f%%")
plt.axis("equal")
plt.title("Female-Male Pie Diagram")
plt.show()
```

```python
data_female = data[data["gender"] == "female"]
data_male = data[data["gender"] == "male"]

female_math_mean = data_female["math score"].mean()
male_math_mean = data_male["math score"].mean()

female_reading_mean = data_female["reading score"].mean()
male_reading_mean = data_male["reading score"].mean()

female_writing_mean = data_female["writing score"].mean()
male_writing_mean = data_male["writing score"].mean()
```

```python
female_ratio_math = female_math_mean / (female_math_mean + male_math_mean) * 100
male_ratio_math = male_math_mean / (female_math_mean + male_math_mean) * 100

labels = ["Female Math Ratio","Male Math Ratio"]
colors = ["pink", "lightblue"]
sizes = [female_ratio_math, male_ratio_math]
plt.pie(sizes, colors = colors, labels = labels, startangle = 90, autopct = "%1.1f%%")
plt.axis("equal")
plt.title("Female-Male Math Ratio Pie Diagram")
plt.show()
```

```python
female_ratio_reading = female_reading_mean / (female_reading_mean + male_reading_mean) * 100
male_ratio_reading = male_reading_mean / (female_reading_mean + male_reading_mean) * 100

labels = ["Female Reading Ratio","Male Reading Ratio"]
colors = ["pink", "lightblue"]
sizes = [female_ratio_reading, male_ratio_reading]
plt.pie(sizes, colors = colors, labels = labels, startangle = 90, autopct = "%1.1f%%")
plt.axis("equal")
plt.title("Female-Male Reading Ratio Pie Diagram")
plt.show()
```

```python
female_ratio_writing = female_writing_mean / (female_writing_mean + male_writing_mean) * 100
male_ratio_writing = male_writing_mean / (female_writing_mean + male_writing_mean) * 100

labels = ["Female Writing Ratio","Male Writing Ratio"]
colors = ["pink", "lightblue"]
sizes = [female_ratio_writing, male_ratio_writing]
plt.pie(sizes, colors = colors, labels = labels, startangle = 90, autopct = "%1.1f%%")
plt.axis("equal")
plt.title("Female-Male Writing Ratio Pie Diagram")
plt.show()
```
